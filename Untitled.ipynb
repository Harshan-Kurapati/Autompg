# Train Test Split

```python
df_new = df.drop("car_name", axis = 1)

X = df_new.drop("mpg", axis = 1)
y = df["mpg"]
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.8, shuffle = True, random_state= 42)


print(f"shape of X_train: {X_train.shape}")
print(f"shape of X_test: {X_test.shape}")
print(f"shape of y_train: {y_train.shape}")
print(f"shape of y_test: {y_test.shape}")
```

# Standardisation

```python
def standardise(x):
    standardscalar = StandardScaler()
    x = x.copy()
    x.loc[:,:] = standardscalar.fit_transform(x)
    
    return x

X_train_S = standardise(X_train)
X_test_S = standardise(X_test)
```

# Linear Regression

```python
X_train_S["bias"] = 1



theta = np.zeros(X_train_S.shape[1])

alpha = 0.01
Iteration = 500
m = len(X_train_S)  

cost_list = [] 
Ite = []

for i in range(Iteration):
    
    h_theta = np.dot(X_train_S,theta)
    
    cost = (1/(2*m))*(np.sum((h_theta - y_train)**2))
    
    J_theta = (1/m) * (np.dot(X_train_S.transpose(), (h_theta - y_train)))
    
    theta -= alpha*J_theta
    
    cost_list.append(cost)
    Ite.append(i)  
    
    
MSE = (1/m)
    
    
     
```
